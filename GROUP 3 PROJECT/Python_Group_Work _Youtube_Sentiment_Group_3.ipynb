```python
#Sort by the years--  Bringing all the Latest Years 
dup_data=dup_data.sort_values(by='year',ascending=False)

#Resetting index after sorting 
dup_data=dup_data.reset_index(drop=True) #Drop == True deletes the older index

#Checking Duplicates based on the primary key (video_id)
dup_data[dup_data['video_id'].duplicated()]
```

```python
#Removing duplicates based on video_id column and keeping the first data
dup_data=dup_data.drop_duplicates(subset='video_id',keep='first')
dup_data
```

```python
dup_data.shape
```

```python
# Changing Column names

new_names=["Name","Trailer Link","Video ID","Sentiment Scores","Favorability","Rating","Genre","Year",
           "Released","Votes","Director","Writer","Star","Prod. Country","Budget ($M)","Gross ($M)","Company","Runtime (min)"]

dup_data.columns =new_names
```

```python
dup_data
```

```python
#converting budget and gross(columns) to million
#Run only once!

dup_data["Budget ($M)"] = round((dup_data["Budget ($M)"]/1000000),2)
dup_data["Gross ($M)"] = round((dup_data["Gross ($M)"]/1000000),2)
dup_data
```

```python
#Creating Returns column.

dup_data["Return ($M)"]=dup_data["Gross ($M)"] -dup_data["Budget ($M)"]
```

```python
dup_data
```

```python
# Splitting Released into Country and the date
dup_data[['Released Date', 'Released Country']] = dup_data['Released'].str.split(('('),expand=True)
dup_data['Released Country'] = dup_data['Released Country'].str.strip(')')

dup_data
```

```python
#dropping the redundant released columns
dup_data = dup_data.drop(columns = ['Released'])

dup_data
```

```python
#removing the curly brackets
dup_data['Sentiment Scores'] = dup_data['Sentiment Scores'].str.strip('{}')

dup_data
```

```python
#splitting one column(sentiment scores) into three new columns(positive,neutral,negative)

dup_data[['Positive Score', 'Neutral Score', 'Negative Score']] = dup_data['Sentiment Scores'].str.split((','),expand=True)

dup_data
```

```python
# dropping the now redundant sentiment score column
dup_data = dup_data.drop(columns = ['Sentiment Scores'])

dup_data
```

```python
#extracting only the digits in the newly created columns

dup_data['Positive Score'] = dup_data['Positive Score'].str.extract(r'(\d+)')
dup_data['Neutral Score'] = dup_data['Neutral Score'].str.extract(r'(\d+)')
dup_data['Negative Score'] = dup_data['Negative Score'].str.extract(r'(\d+)')

dup_data
```

```python
#creating a new column that checks if a movie was profitable

dup_data['Profit/Loss']=dup_data['Return ($M)'].apply(lambda x:'Loss' if x < 0  else ('Break-even' if x == 0 else 'Profit') )

dup_data
```

```python
#checking the original column order

og_file.columns
```

```python
#checking our current column order
dup_data.columns
```

```python
# creating a list for our rearranged columns
rearranged = ['Name', 'Trailer Link', 'Video ID', 'Positive Score', 'Neutral Score',
       'Negative Score', 'Favorability', 'Rating', 'Genre',
       'Year', 'Released Date', 'Released Country' ,'Votes', 'Director', 'Writer', 'Star', 'Prod. Country',
       'Budget ($M)', 'Gross ($M)', 'Return ($M)', 'Profit/Loss','Company', 'Runtime (min)']
```

```python
#reindexing our columns
dup_data = dup_data.reindex(columns= rearranged)

dup_data
```

```python
#rounding our favourability metrics to 3 decimal places
dup_data['Favorability'] = round(dup_data['Favorability'], 3)

dup_data
```

```python
#displaying current data types
dup_data.info()
```

```python
#data types we will be changing

#3 sentiment scores - int
#rating & genre - category
#date to data
#released country,production country,company,profit/loss - category ?
```

```python
#changing all the necessary data types 

dup_data['Positive Score'] = dup_data['Positive Score'].astype('int')
dup_data['Neutral Score'] = dup_data['Neutral Score'].astype('int')
dup_data['Negative Score'] = dup_data['Negative Score'].astype('int')
dup_data['Released Date'] = dup_data['Released Date'].astype('datetime64[ns]')
dup_data['Rating'] = dup_data['Rating'].astype('category')
dup_data['Genre'] = dup_data['Genre'].astype('category')
dup_data['Released Country'] = dup_data['Released Country'].astype('category')
dup_data['Prod. Country'] = dup_data['Prod. Country'].astype('category')
dup_data['Company'] = dup_data['Company'].astype('category')
dup_data['Profit/Loss'] = dup_data['Profit/Loss'].astype('category')
dup_data.info()
```

```python
#checking to see where we still have null values
dup_data.isnull().sum()
```

```python
#dropping rows with nulls in their Budget column
dup_data = dup_data.dropna(subset= ['Budget ($M)'])

#confirming nulls from budget column have been dropped
dup_data.isnull().sum()
```

```python
#resetting indices after removing nulls

dup_data=dup_data.reset_index(drop=True) #Drop == True deletes the older index
```

```python
#calling the row with null runtime value

dup_data[dup_data['Runtime (min)'].isna()]
```

```python
#populating null cell in Runtime column (value obtained from google search)

dup_data.fillna(value= {'Runtime (min)' : 91}, inplace = True)
```

# VISUALIZATION

```python
# Question: Looking at our data, what do you want us to visualize (chart)?
# Top 5 movies based on their Returns (Column Chart)
# Rating and Gross Visualization (Column)
# Count of Genre from the movie list
# 10 Directors and budget
# Top 5 Star-Return (Bar)
# Top 7 countries where movies were produced
# Percentage of movies that made profits and loss/ Breakeven.
```

```python
#Creating Pivot Tables.
# Top 5 movies based on their Returns (Column Chart)
top_5=clean_data.groupby(by='Name')['Return ($M)'].sum().sort_values(ascending=False).to_frame().head(5)
top_5.sort_values(by='Return ($M)',inplace=True)
top_5

```

```python
#the chart

plt.barh(top_5.index,top_5['Return ($M)'], color=['#FF7F7F'])
plt.xlabel("Return ($M)")
plt.ylabel("Movies")
plt.title("Top 5 Movies Per Returns")
plt.savefig('top_5_by_returns.png')  #saving our figure
plt.show()
```

```python
# Rating and Gross Visualization (Column)
rating_avggross=round((clean_data.groupby(by='Rating')['Gross ($M)'].mean().sort_values(ascending=False).to_frame()),2)
rating_avggross
```

```python
# the chart

plt.figure(figsize=(15,8))
plt.bar(rating_avggross.index,rating_avggross['Gross ($M)'],color = ['#FF5733','#FFC300','#702963','#CC5500','#D70040', '#954535', '#F88379', '#986868'])
plt.xlabel("Rating")
plt.xticks(fontsize=8) #Helps reduce the font size of the x-axis
plt.ylabel("Avg Gross ($/M)")
plt.title("Average Gross For Each Rating")
plt.savefig('avg_gross_by_rating.png') #saving our figure
plt.show()
```

```python
# Count of Genre from the movie list

genre_count=clean_data.groupby(by='Genre')['Name'].count().sort_values(ascending=False).to_frame()
genre_count
```

```python
#the chart

plt.figure(figsize=(15,8))
plt.bar(genre_count.index,genre_count['Name'], color = ['#FF5733','#FFC300','#702963','#CC5500','#D70040', '#954535', '#F88379', '#986868'])
plt.xlabel("Genre")
plt.xticks(fontsize=10) #Helps reduce the font size of the x-axis
plt.ylabel("Number of Movies")
plt.title("Number of Movies in each Genre")
plt.savefig('num_movies_by_genre.png') #saving our figure
plt.show()
```

```python
# 10 Directors and avg budget
top10_dir=round((clean_data.groupby(by='Director')['Budget ($M)'].mean().sort_values(ascending=False).to_frame().head(10)),2)
top10_dir.sort_values(by='Budget ($M)',inplace=True)
top10_dir
```

```python
#the chart

#plt.figure(figsize=(8,8))
plt.barh(top10_dir.index,top10_dir['Budget ($M)'],color=['#FF7F7F'])
plt.xlabel("Budget ($M)")
plt.ylabel("Directors")
plt.title("Top 10 Directors vs Average Budget")
plt.savefig('top_directors_by_avgbudget.png') #saving our figure
plt.show()
```

```python
# Top 5 Star- Avg Return (Bar)
star_returns=round((clean_data.groupby(by='Star')['Return ($M)'].mean().sort_values(ascending=False).to_frame().head(5)),2)
star_returns
```

```python
#the chart

plt.figure(figsize=(12,8))
plt.bar(star_returns.index,star_returns['Return ($M)'],color=['#FF7F7F'])
plt.xlabel("Stars")
plt.ylabel("Average Returns($M)")
plt.title("Top 5 Star With Highest Average Returns")
plt.savefig('top_stars_by_returns.png') #saving our figure
plt.show()
```

```python
# Top 7 countries where movies were produced
top7_country=clean_data.groupby(by='Prod. Country')['Prod. Country'].count().sort_values(ascending=False).to_frame().head(7)
top7_country
```

```python
#the chart

plt.figure(figsize=(10,8))
plt.bar(top7_country.index,top7_country['Prod. Country'],color = ['#FF5733','#FFC300','#702963','#CC5500','#D70040', '#954535', '#F88379', '#986868'])
plt.xlabel("Budget ($M)")
plt.ylabel("Directors")
plt.title("Top 7 Countries with Highest Production")
plt.xticks(fontsize=8)
plt.savefig('top_countries_by_budget.png') #saving our figure
plt.show()
```

```python
# Percentage of movies that made profits and loss/ Breakeven

ratio_lbp=clean_data.groupby(by='Profit/Loss')['Profit/Loss'].count().to_frame()
ratio_lbp
```

```python
#the chart

colors_list=['red','gold','green',]

explode_list= [0,0.2,0]

plt.pie(ratio_lbp["Profit/Loss"], 
        startangle = 90,
        shadow=True,
        autopct = '%.1f%%',
        labels=ratio_lbp.index,
       colors=colors_list,
        explode=explode_list)

plt.title('Proportion of Profitability by Movie')
plt.savefig('pie_chart.png') #saving our figure
plt.show()
```

```python
#Year on Year Sum Returns (Line Chart)

YoY_Returns=clean_data.groupby(by='Year')['Return ($M)'].mean().to_frame()
YoY_Returns
```

```python
#the chart

plt.figure(figsize=(15,8))
plt.plot(YoY_Returns.index,YoY_Returns['Return ($M)'],c='r', marker="o")
plt.xlabel("Years")
plt.ylabel("Average Return ($M)")
plt.title("Trend of Average Returns (2005-2019)")
plt.xticks(fontsize=8)
plt.savefig('line_chart.png') #saving our figure
plt.show()
```
