# Algorithim for Data cleaning
- checking for and removing duplicates
- Check and remove null values
- remove unneeded columns
- renaming the columns 
- change birthdate data type frpm object to date
- importing dateframe for the date
- converting non-numeric values to numeric

```python
#checking for and removing duplicates
billionaire.duplicated()
```

```python
billionaire[billionaire.duplicated()]
```

```python
#Checking for number of  duplicate entries
billionaire.duplicated().sum()
```

```python
drop_columns = ['lastName','firstName','birthYear','birthMonth','birthDay','gross_tertiary_education_enrollment','gross_primary_education_enrollment_country','latitude_country','longitude_country','date','residenceStateRegion','cpi_change_country','life_expectancy_country','category','cpi_country','population_country']
```

```python
billionaire.drop(drop_columns, axis = 1, inplace=True)
```

```python
billionaire
```

```python
# renaming the columns
billionaire_rename = ['Rank', 'Networth($/B)', 'Name', 'Age', 'Country', 'City', 'Source',
                      'Industries', 'Nationality', 'Organization', 'Selfmade','Status', 
                      'Gender', 'Birthdate', 'Title', 'State','GDPCountry','TaxRevenue','TaxRate']
```

```python
billionaire.columns = billionaire_rename
```

```python
billionaire.head()
```

```python
# checking for null values
billionaire.isnull().sum()
```

```python
billionaire.fillna(value= {'Organization' : "Unknown",'Title' : "Unknown",'State' : "Unknown",'Country' : "Unknown",'City' : "Unknown"}, inplace = True)
```

```python
billionaire.isnull().sum()
```

```python
billionaire.shape
```

```python
#Removing zeros from Networth column
billionaire["Networth($/B)"]=round((billionaire["Networth($/B)"]/1000),0)
billionaire
```

```python
#checking for data types 
billionaire.dtypes
```

```python
#converting the GDPCountry from an object to float
#converting non-numeric values to numeric
billionaire["GDPCountry"] = pd.to_numeric(billionaire["GDPCountry"], errors = 'coerce')
billionaire["GDPCountry"].dtype
```

```python
#changing birthdate type to date
billionaire['Birthdate'] = pd.to_datetime(billionaire['Birthdate'])
billionaire['Birthdate'].dtype
```

```python
#changing age data type to integer
billionaire['Age'] = billionaire['Age'].fillna(-1).astype(int)
billionaire['Age'].dtype
```

```python
billionaire.dtypes
```

```python
# billionaire with highest networth
highest_networth = billionaire.groupby("Name")["Networth($/B)"].sum().sort_values(ascending = False).to_frame().head(5)
highest_networth
```

```python
# billionaire with the highest networth
highest_networth = billionaire.groupby("Name")["Networth($/B)"].sum().sort_values(ascending = False).to_frame().head(5)
highest_networth
plt.figure(figsize=(10,8))
plt.bar(highest_networth.index, highest_networth["Networth($/B)"])
plt.xlabel("Name")
plt.xticks(fontsize=10)
plt.ylabel("Networth")
plt.title("Top 5 Billionaires")
print(plt.show())
```

```python
gender_count = billionaire["Gender"].value_counts().to_frame()
gender_count
```

```python
# percentage of male and female billionares 
plt.pie(gender_count["Gender"],autopct = '%.1f%%',labels=gender_count.index);
```

```python
#country with the most billionaires
country_billionaires = billionaire["Country"].value_counts().to_frame()
country_billionaires
```

```python
# countries with the most billionaires
country_billionaires = billionaire["Country"].value_counts().to_frame().head(10)
country_billionaires

plt.figure(figsize=(15,8))
plt.bar(country_billionaires.index, country_billionaires["Country"])
plt.xlabel("Countries")
plt.xticks(fontsize=10)

plt.ylabel("Billionaires_count")
plt.title("Country with the most Billionaires(10)")
print(plt.show())
```

```python
# industries vs networth
industries_networth = billionaire.groupby("Industries")["Networth($/B)"].sum().sort_values(ascending = False).to_frame()
industries_networth
plt.figure(figsize=(10,8))
plt.hist(x=billionaire["Industries"], bins=10)
plt.xlabel("Industries")
plt.xticks(fontsize=10, rotation=90)
plt.ylabel("Networth")
plt.title("Histogram of Industries vs Networth")
print(plt.show())
```
